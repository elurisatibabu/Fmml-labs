## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
## TASK-5 : Run gradient descent 3 times
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Gradient descent parameters
learning_rate = 0.1
iterations = 10

# Run gradient descent three times with different starting points
starting_points = [5, -5, 2]
results = []

for start in starting_points:
    x = start
    trajectory = [x]
    for _ in range(iterations):
        x = x - learning_rate * df(x)
        trajectory.append(x)
    results.append(trajectory)

# Plot the results
x_vals = np.linspace(-6, 6, 400)
y_vals = f(x_vals)

plt.figure(figsize=(12, 6))

for i, trajectory in enumerate(results):
    plt.plot(x_vals, y_vals, label='f(x) = x^2')
    plt.plot(trajectory, f(np.array(trajectory)), marker='o', label=f'Starting point {starting_points[i]}')
    plt.title('Gradient Descent Trajectories')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

## TASK-6 : Plot two more learning rates: mid_gamma (0.001) and high_gamma (0.002)
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Gradient descent parameters
iterations = 10
learning_rates = [0.1, 0.001, 0.002]
starting_point = 5

# Run gradient descent with different learning rates
results = []

for lr in learning_rates:
    x = starting_point
    trajectory = [x]
    for _ in range(iterations):
        x = x - lr * df(x)
        trajectory.append(x)
    results.append((lr, trajectory))

# Plot the results
x_vals = np.linspace(-6, 6, 400)
y_vals = f(x_vals)

plt.figure(figsize=(18, 6))

for i, (lr, trajectory) in enumerate(results):
    plt.subplot(1, 3, i+1)
    plt.plot(x_vals, y_vals, label='f(x) = x^2')
    plt.plot(trajectory, f(np.array(trajectory)), marker='o', label=f'Learning rate {lr}')
    plt.title(f'Gradient Descent with Learning Rate {lr}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()

plt.tight_layout()
plt.show()

````

````python
## TASK-7 : Plot reduction in cost for each iteration
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Gradient descent parameters
learning_rates = [0.1, 0.001, 0.002]
iterations = 50
starting_point = 5

# Run gradient descent with different learning rates and track the cost
results = []

for lr in learning_rates:
    x = starting_point
    costs = []
    for _ in range(iterations):
        cost = f(x)
        costs.append(cost)
        x = x - lr * df(x)
    results.append((lr, costs))

# Plot the reduction in cost for each iteration
plt.figure(figsize=(12, 6))

for lr, costs in results:
    plt.plot(costs, label=f'Learning rate {lr}')

plt.title('Reduction in Cost for Each Iteration')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.legend()
plt.show()
````

````python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
````

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
