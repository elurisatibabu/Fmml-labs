# **Section 1: Tweaking the parameters**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
```

## Section 1.1: Changing Value of K
Let's try to run the knn with different values of knn and compare the results

```python
neighbors = []
scores_list = []

for k in range(1, 25):
    neighbors.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train,y_train)
    scores = knn.score(X_test,y_test)
    scores_list.append(scores)
```

```python
# determining the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print(max_acc)
print(optimal_ks)
```

### AUC-ROC Curve

The AUC - ROC curve is a performance measurement for classification problems at various threshold settings. ROC is a probability curve and AUC represents the degree or measure of separability.
It tells how much the model is capable of distinguishing between classes.  
The diagonal in the curve divides the ROC space. Points above the diagonal represent good classification results (better than random); points below the line represent bad results (worse than random).    
Higher the AUC, the better the model is at predicting 0 classes as 0 and 1 classes as 1.

The ROC curve is plotted with the TPR (True Positive Rate) on the y-axis, and the FPR (False Positive Rate) on the x-axis.

```python
## AUC-ROC
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pyplot as plt

def plot_roc(y_test,predicted):
  y_scores = predicted
  fpr, tpr, threshold = roc_curve(y_test, y_scores)
  roc_auc = auc(fpr, tpr)
  plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
  plt.legend(loc = 'lower right')
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0, 1])
  plt.ylim([0, 1])
  plt.ylabel('True Positive Rate')
  plt.xlabel('False Positive Rate')
  plt.title('Receiver Operating Characteristic')
  plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train,y_train)
predictions = knn.predict(X_test)
plot_roc(y_test,predictions)
```

# **Section 2: Cross Validation**

As you saw in the above two cases with different train-test splits, there is a risk of overfitting on the test set because the parameters can be tweaked until the estimator performs optimally. This way, knowledge about the test set can “leak” into the model and evaluation metrics no longer report on generalization performance.  

Cross Validation is an important technique in Machine Learning to analyze the performance of a model on **unseen** data.  

There are different kinds of cross-validation methods: exhaustive and non-exhaustive.   
Exhaustive cross-validation methods are cross-validation methods which learn and test on all possible ways to divide the original sample into a training and a validation set.
Non-exhaustive cross validation methods do not compute all ways of splitting the original sample.

One of the most commonly cross-validation methods is **K-Fold Cross validation**. The procedure for K-Fold Cross Validation is as follows:  
1. Split the dataset into k groups (typically without shuffling)
2. For each unique group:  
  a. Take the group as a hold out or test data set  
  b. Take the remaining groups as a training data set  
  c. Fit a model on the training set and evaluate it on the test set  
  d. Retain the evaluation score and discard the model  
3. Summarize the performance of the model using the mean of model evaluation scores over all the folds.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 17)
```

```python
from sklearn.model_selection import KFold
from statistics import mean
kf = KFold(n_splits=10)
cv_scores = []
neighbors = []
for k in range(1,25):
  knn = KNeighborsClassifier(n_neighbors = k)
  neighbors.append(k)
  acclist = []
  for train_index, test_index in kf.split(X_train, y_train):
    x_train_fold, x_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    knn.fit(x_train_fold, y_train_fold)
    acclist.append(knn.score(x_test_fold, y_test_fold))
  cv_scores.append(mean(acclist))
```

```python
Accuracy = [x*100 for x in cv_scores]

# plot misclassification error versus k
plt.figure(figsize = (10, 6))
plt.plot(neighbors, Accuracy)
plt.xlabel('Number of neighbors')
plt.ylabel('Mean Cross-validation accuracy in %')
plt.show()
```

```python
# determining the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print(optimal_ks)
```

Note that the accuracy in the above case is not the accuracy on the test set. We plot the mean of the accuracies obtained during cross-validation, when we considered different splits of the training set.  

From the plot above, we can see that when the value of k is 10 or 12, our KNN model generalizes well to unseen data. Performing such cross validation gives us an idea about the parameters to choose when using a machine-learning algorithm.

# **Section 3: Standardization**


Scaling the data that we have available is a recommended pre-processing step when working with many machine learning algorithms. The scale and distribution of the data drawn from the domain may be different for each variable. Differences in the scales across input variables may increase the difficulty of the problem being modeled, and the model may suffer from poor performance during learning and sensitivity to input values resulting in higher generalization error.  

Standardization (also called z-score normalization) is one such scaling technique that transforms your data such that the resulting distribution has a mean of 0 and a standard deviation of 1.

$$ \text{New (scaled) value} = {X - \mu \over \sigma}$$

1. Instead of using cross-validation, we could also split our dataset into three sets- train, validation, and test set. The training proceeds on the training set, after which evaluation is done on the validation set, and when the experiment seems to be successful, final evaluation can be done on the test set.  
What do you think are the pros/cons of this method ?
**Answer
  Splitting the dataset into three distinct sets—training, validation, and test—has its own advantages and disadvantages. Here’s a breakdown of the pros and cons of this approach:

Pros

1. Clear Evaluation Phases:

Training Set: Used to fit the model, allowing it to learn patterns.

Validation Set: Helps in tuning hyperparameters, selecting the best model, and preventing overfitting. This separation ensures that the model evaluation during tuning is unbiased, as the model has not seen this data.

Test Set: Provides a final, unbiased evaluation of the model's performance after training and validation. This is crucial for assessing the model's generalization to unseen data.



2. Reduced Overfitting:

By using a separate validation set, you can monitor the model's performance on unseen data during training. This helps identify overfitting early, as performance on the validation set may degrade while performance on the training set improves.



3. Better Hyperparameter Tuning:

The validation set allows for systematic experimentation with different hyperparameters, model architectures, or feature selections without contaminating the final performance evaluation.



4. Model Selection:

You can evaluate multiple models or configurations based on their performance on the validation set, choosing the best one for the final test.




Cons

1. Data Utilization:

Dividing the dataset into three subsets reduces the amount of data available for training. This can be particularly problematic with small datasets, as the model may not learn well if it has insufficient data.



2. Variance in Performance Metrics:

The performance on the validation set can vary depending on the random split of the data. This means that results may not be as stable or reliable as those obtained through techniques like cross-validation, where multiple splits provide a more robust estimate of performance.



3. Model Dependency:

If the validation set is not representative of the overall dataset, it could lead to suboptimal hyperparameter tuning or model selection. This can happen especially in cases of imbalanced datasets or small sample sizes.



4. Risk of Overfitting to Validation Set:

There's a risk of overfitting the model to the validation set if it is used excessively for tuning. This can lead to an inflated estimate of model performance when evaluated on the test set.




Summary

Using a three-way split (training, validation, and test) can be a good strategy, especially in situations where model tuning is necessary and where enough data is available to maintain the integrity of each subset. However, care must be taken to ensure that the validation set remains representative and that it does not lead to overfitting. For smaller datasets, alternative methods such as cross-validation might provide more reliable estimates of model performance without sacrificing data availability.



2. Is K-Fold cross validation an exhaustive or non-exhaustive cross validation method ? Justify your answer.

**Answer
K-Fold Cross Validation is considered a non-exhaustive cross-validation method. Here’s the justification for this classification:

Explanation of K-Fold Cross Validation

1. Process: In K-Fold Cross Validation, the dataset is divided into  subsets (or folds). The model is trained  times, each time using  folds for training and the remaining fold for validation. This process allows each data point to be used for validation exactly once.


2. Exhaustive vs. Non-Exhaustive:

Exhaustive Methods: These methods evaluate every possible combination or configuration. For instance, in exhaustive cross-validation, every possible training and validation split would be evaluated, which could become computationally prohibitive, especially with larger datasets. This approach includes techniques like Leave-One-Out Cross-Validation (LOOCV), where each instance in the dataset serves as a validation set while the rest form the training set.

Non-Exhaustive Methods: K-Fold Cross Validation does not evaluate every possible combination of training and validation sets. Instead, it takes a fixed number of folds (the value of ) and partitions the data into those  subsets. Each subset is used exactly once for validation, and thus the process does not cover all possible splits of the dataset.




Justification for K-Fold being Non-Exhaustive

Fixed Number of Iterations: K-Fold Cross Validation operates with a predetermined number of folds, and the model is trained and validated a fixed number of times (specifically  times). This limited approach contrasts with exhaustive methods, which consider every conceivable combination of training and validation data.

Efficiency: The method is computationally more efficient compared to exhaustive methods. By not considering all combinations, it significantly reduces the computational burden, especially for large datasets.

Generalization: While K-Fold Cross Validation provides a good estimate of the model’s performance, it does not guarantee that every potential combination of training/validation splits is tested. This allows for quicker assessments while still offering insights into model stability and generalization.


Conclusion

In summary, K-Fold Cross Validation is a non-exhaustive method because it does not evaluate every possible training/validation split. Instead, it systematically partitions the data into  folds, providing a balance between computational efficiency and robust performance estimation.
